```python
import numpy as np 
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation

```

```python
pos_point=np.array([(-2,3),(0,1),(2,-1)])
neg_point=np.array([(-2,1),(0,-1),(2,-3)])
print(pos_point)
# print(neg_point)
```

```python
plt.figure(figsize=(8,6))
plt.scatter(pos_point[:,0],pos_point[:,1],color='blue',label='positive class',marker='o')
plt.scatter(neg_point[:,0],neg_point[:,1],color='red',label='negative class',marker='x')
plt.xlabel('x1')
plt.ylabel('x2')
plt.legend()
plt.title('Data Points')
plt.grid(True)
plt.show()
```

```python
X=np.vstack((pos_point,neg_point))
y=np.array([1,1,1,-1,-1,-1])
```

```python
def perceptron(X,y,steps=10):
    theta=np.zeros(X.shape[1])
    thera0=0
    for t in range(steps):
        for i,xi in enumerate(X):
            if y[i]*(np.dot(theta,xi)+thera0)<=0:
                theta+=y[i]*xi
                thera0+=y[i]
                break
        plot_decision_boundary(theta,thera0,t+1)
    return theta,thera0
        
```

```python
def plot_decision_boundary(theta,thera0,iteration):
    plt.figure(figsize=(8,6))
    plt.scatter(pos_point[:,0],pos_point[:,1],color='blue',label='positive class',marker='o')
    plt.scatter(neg_point[:,0],neg_point[:,1],color='red',label='negative class',marker='x')
    x=np.linspace(-3,3,100)
    if theta[1]!=0:
        y=-(theta[0]*x+thera0)/theta[1]
        plt.plot(x,y,color='green',label=f'iteration {iteration}')
    else:
        x_val=-thera0/theta[0]
        plt.axvline(x=x_val,color='green',label=f'iteration {iteration}')
    plt.xlabel('x1')
    plt.ylabel('x2')
    plt.legend()
    plt.title('Perceptron Decision Boundary')
    plt.grid(True)
    plt.xlim(-3,3)
    plt.ylim(-4,4)
    plt.show()
```

```python
theta_final,thera0_final=perceptron(X,y,steps=10)
```
