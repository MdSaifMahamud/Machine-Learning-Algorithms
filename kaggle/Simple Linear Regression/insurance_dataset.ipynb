```python
import pandas as pd 
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score,root_mean_squared_error
```

```python
df=pd.read_csv('simplelinearregression.csv')
df
```

```python
plt.scatter(df['Age'],df['Premium'])
```

```python
model=LinearRegression()
x_train=df['Age'].values.reshape(-1,1)
y_train=df['Premium'].values
model.fit(x_train,y_train)
```

```python
x_end=[x_train.min(),x_train.max()]
plt.scatter(x=x_train,y=y_train,color='red',alpha=0.6,label='Train')
# plt.scatter(x=x_test,y=y_test,color='b',alpha=0.6,label='Test')
print(model.coef_)
print(model.intercept_)
y_end=model.coef_ * x_end + model.intercept_
plt.plot(x_end,y_end,label='Line')
plt.legend()
plt.show()
```
