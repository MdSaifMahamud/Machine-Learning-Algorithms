```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data=pd.read_csv('homeprices.csv')
plt.scatter(data.area, data.price, color='red', marker='x')
plt.show()
```

```python
def loss_function(m,b,points):
    total_error=0
    for i in range(len(points)):
        x=points.iloc[i].area
        y=points.iloc[i].price
        total_error += (y - (m*x + b))**2
    return total_error / len(points)
```

```python
def gradient_descent(m_now,b_now,points,L):
    m_grad=0
    b_grad=0
    n=len(points)
    for i in range(n):
        x=points.iloc[i].area
        y=points.iloc[i].price
        m_grad += -(2/n) * x * (y - (m_now*x + b_now))
        b_grad += -(2/n) * (y - (m_now*x + b_now))
    m= m_now - L * m_grad
    b= b_now - L * b_grad
    return m, b
```

```python
m=0
b=0
L=.000000001
epochs= 1000
for i in range(epochs):
    m, b = gradient_descent(m, b, data, L)
    print(f"Epoch {i+1}: m={m}, b={b}, loss={loss_function(m, b, data)}")
print("Final m:", m, "Final b:", b)
plt.scatter(data.area, data.price, color='black')
plt.plot(list(range(2600,4000)),[m*x+b for x in range(2600,4000)], color='red')
plt.show()
    
```
