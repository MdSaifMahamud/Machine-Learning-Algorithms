```python
import numpy as np 
import matplotlib.pyplot as plt
```

```python
x_train=np.array([1.0,2.0])
y_train=np.array([300.0,500.0])
print(x_train)
print(y_train)
```

```python
plt.scatter(x_train,y_train,marker='*',c='r')
plt.xlabel('x_train')
plt.ylabel('y_train')
plt.show()
```

```python
w=100
b=100
```

```python
def model_output(x,w,b):
    m=x.shape[0]
    f_wb=np.zeros(m)
    for i in range(m):
        f_wb[i]=w*x[i]+b
    return f_wb
    
```

```python
tmp_f_wb=model_output(x_train,w,b)
print(tmp_f_wb)
plt.plot(x_train,tmp_f_wb,c='b')
plt.scatter(x_train,y_train,marker='*',c='r')
plt.xlabel('x_train')
plt.ylabel('y_train')
plt.legend(['Model Output','Training Data'])
plt.show()
```
